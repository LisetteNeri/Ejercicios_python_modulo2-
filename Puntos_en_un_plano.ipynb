```python
# Importamos la función hypot del módulo math
from math import hypot

class Point:
    def __init__(self, x=0.0, y=0.0):
        # Coordenadas privadas
        self.__x = float(x)
        self.__y = float(y)

    def getx(self):
        # Devuelve la coordenada x
        return self.__x

    def gety(self):
        # Devuelve la coordenada y
        return self.__y

    def distance_from_xy(self, x, y):
        # Calcula la distancia entre el punto actual y otro punto (x, y)
        return hypot(self.__x - x, self.__y - y)

    def distance_from_point(self, point):
        # Calcula la distancia entre el punto actual y otro objeto Point
        return hypot(self.__x - point.getx(), self.__y - point.gety())
        # Creamos dos puntos
p1 = Point(0, 0)
p2 = Point(1, 1)

# Mostramos coordenadas
print("Coordenadas de p1:", p1.getx(), p1.gety())
print("Coordenadas de p2:", p2.getx(), p2.gety())

# Distancia entre p1 y coordenadas (0,0)
print("Distancia de p1 a (0,0):", p1.distance_from_xy(0, 0))

# Distancia entre p1 y p2
print("Distancia de p1 a p2:", p1.distance_from_point(p2))
```
